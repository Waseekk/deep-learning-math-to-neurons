## Single Neuron
A neuron is the building block of neural networks! It takes inputs, multiplies by weights, adds bias, and applies an activation function.

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
inputs=torch.tensor([2,3]) #input values for the neuron. this can be a vector of any size depending on the number of inputs to the neuron. mathmatically, this is often represented as a column vector. for example, if we have 3 inputs, it would be a 3x1 vector like [[x1], [x2], [x3]]. it can also be represented as a 1D array like [x1, x2, x3] for simplicity in code. the choice of representation depends on the context and the operations being performed. in this case, we are using a 1D tensor for simplicity.
weights=torch.tensor([0.5,0.3]) #weights for each input, these can be learned during training. initially, they can be set to random values or small constants. the weights determine the importance of each input in the neuron's output. for example, if we have 3 inputs, we would have 3 corresponding weights like [w1, w2, w3]. the weighted sum is calculated as w1*x1 + w2*x2 + w3*x3. in this case, we have 2 inputs and 2 weights.
bias=.1 #bias is an additional parameter that allows the neuron to shift the activation function. it can be thought of as the neuron's threshold for firing. the bias can also be learned during training and is often initialized to zero or a small constant. in this case, we have set it to 0.1.

weighted_sum=inputs[0]*weights[0]+inputs[1]*weights[1]+bias #calculate the weighted sum of the inputs and bias. this is the linear combination of the inputs and weights plus the bias. mathematically, it can be represented as z = w1*x1 + w2*x2 + ... + wn*xn + b, where z is the weighted sum, w are the weights, x are the inputs, and b is the bias. in this case, we have 2 inputs and 2 weights, so it simplifies to z = w1*x1 + w2*x2 + b. axis
print(f"Weighted sum = {weights[0]} × {inputs[0]} + {weights[1]} × {inputs[1]} + {bias}")
print(f"Weighted sum = {weighted_sum:.2f}")
```

```python
# Apply activation (sigmoid)
activation = 1 / (1 + torch.exp(-weighted_sum))
print(f"After activation (sigmoid): {activation:.4f}")
print()
print("This is how a single neuron works!")
```

## Step 1: Initialize a Neuron

```python
class Neuron:
    "step-1: Intiailize the neuron with weights and bias"

    def __init__(self, num_inputs):
        self.weights=torch.randn(num_inputs)*.1 #randomly initialize weights with small values. 
                                                #this is important to break symmetry during training. if all weights are initialized to the same value, the neuron will learn the same features and will not be able to learn effectively. by initializing with small random values, we allow the neuron to learn different features and improve training. the weights can also be initialized using other methods like Xavier or He initialization for better performance in deeper networks.
        self.bias=torch.tensor(0.0) #initially set to zero. the bias can also be initialized to a small constant or random value, but zero is a common choice for simplicity.
neuron=Neuron(num_inputs=2)
print("num_inputs=2")
print(f"Initial weights: {neuron.weights}")
print(f"Initial bias: {neuron.bias}")
```

## Step 2: Calculate Weighted Sum
 weighted sum (weights · inputs + bias).

```python
class Neuron:
    "Step 2: Add weighted sum calculation"
    def __init__(self, num_inputs):
        self.weights=torch.randn(num_inputs) * .05
        self.bias=torch.tensor(0.0) 

    def weighted_sum(self, inputs):
        " Calculate the weighted sum of the inputs and bias. This is the linear combination of the inputs and weights plus the bias."
# mathematically, it can be represented as z = w1*x1 + w2*x2 + ... + wn*xn + b, where z is the weighted sum, w are the weights, x are the inputs, and b is the bias."
        return torch.dot(self.weights, inputs) + self.bias
  
```

## The Math Behind a Neuron


### The Formula

$$z = \sum_{i=1}^{n} w_i x_i + b$$

Where:
- $z$ = weighted sum (output)
- $w_i$ = weights
- $x_i$ = inputs
- $b$ = bias

---

### Expanded Form

$$z = w_1 x_1 + w_2 x_2 + b$$

---

### Plugging In Our Values

$$w_1 = 0.5, \quad w_2 = 0.3, \quad x_1 = 2.0, \quad x_2 = 3.0, \quad b = 0.1$$

$$z = (0.5 \times 2.0) + (0.3 \times 3.0) + 0.1$$

$$z = 1.0 + 0.9 + 0.1$$

$$\boxed{z = 2.0}$$

---

### What `torch.dot()` Computes

$$\vec{w} \cdot \vec{x} = \sum_{i=1}^{n} w_i x_i = (0.5 \times 2.0) + (0.3 \times 3.0) = 1.9$$

$$z = \vec{w} \cdot \vec{x} + b = 1.9 + 0.1 = 2.0$$

---

### Intuition

| Term | Value | Meaning |
|------|-------|---------|
| $w_1 x_1$ | $0.5 \times 2.0 = 1.0$ | contribution of input 1 |
| $w_2 x_2$ | $0.3 \times 3.0 = 0.9$ | contribution of input 2 |
| $b$ | $0.1$ | baseline shift |
| $z$ | $2.0$ | final output |

> **Key insight:** Weights control *how much* each input matters. Bias shifts the output regardless of inputs.

## STEP 3: aDD ACTIVATION FUNCTION

```python
class Neuron:
    def __init__(self, num_inputs):
        self.weights=torch.randn(num_inputs) * .05
        self.bias=torch.tensor(0.0)
    def weighted_sum(self, inputs):
        return torch.dot(self.weights, inputs) + self.bias
    def sigmoid(self, x):
        "Sigmoid activation function. This function maps any real-valued number into the (0, 1) interval. It is defined as sigmoid(x) = 1 / (1 + exp(-x)). The sigmoid function is commonly used in neural networks to introduce non-linearity and to output values that can be interpreted as probabilities."
        return 1 / (1 + torch.exp(-x))
```

```python
# Test it step by step
neuron = Neuron(num_inputs=2)
neuron.weights = torch.tensor([0.5, 0.3])  # Set fixed weights for demo
neuron.bias = 0.1

inputs = torch.tensor([2.0, 3.0])

# Step 1: Calculate weighted sum
ws = neuron.weighted_sum(inputs)
print("Step 3: Adding Activation Function")
print(f"1. Weighted sum = {ws:.4f}")

# Step 2: Apply sigmoid
output = neuron.sigmoid(ws)
print(f"2. After sigmoid activation = {output:.4f}")
print()
print("Perfect! Now we have a complete neuron with activation.")
```

## 4: Complete Forward Pass

### What a Neuron Does

A neuron takes multiple inputs, scales each by a **weight** (importance),
adds a **bias** (shift), then passes the result through an **activation function** (non-linearity).

---

### Key Components

| Component | Symbol | Role |
|-----------|--------|------|
| Weights | $w_i$ | How important each input is |
| Bias | $b$ | Shifts the weighted sum up or down |
| Activation | $f$ | Makes the neuron non-linear (sigmoid, ReLU, tanh...) |

---

### Forward Pass

**Step 1 — Weighted Sum:**

$$z = \vec{w} \cdot \vec{x} + b = \sum_{i=1}^{n} w_i x_i + b$$

**Step 2 — Apply Activation:**

$$\text{output} = f(z)$$

**Full pipeline:**

$$\text{output} = f\!\left(\sum_{i=1}^{n} w_i x_i + b\right)$$

---

$$x_2 \times w_2 \;\; + \;\; b \;\;\longrightarrow\;\; f(z) \;\;\longrightarrow\;\; \text{output}$$
$$x_n \times w_n \;\; +$$

> Without the activation function, no matter how many layers you stack,
> the network is just a linear transformation — the activation is what gives neural networks their power.

## Single neuron

```python
class Neuron:
    def __init__(self, num_inputs):
        self.weights=torch.randn(num_inputs) * .1
        self.bias=torch.tensor(0.0)
    
    # def weighted_sum(self,inputs):
    #     return torch.dot(self.weights,inputs)+self.bias
    def sigmoid(self,x):
        return 1/1+torch.exp(-x)
    
    def forward(self, inputs):
        weighted_sum= torch.dot(self.weights,inputs)+self.bias
        output=self.sigmoid(weighted_sum)
        return output
    
neuron=Neuron(num_inputs=2)
#neuron.weights=torch.tensor([0.5,0.3]) #manually set weights for testing
neuron.bias=.1

inputs=torch.tensor([2.0,3.0]) #input values for the neuron
output=neuron.forward(inputs)
```

```python
print(f"Weights: {neuron.weights}")
print(f"Bias: {neuron.bias}")
print()
print(f"Output: {output:.4f}")
print()
print("Excellent! Our simplw neuron is working!")
```

## different activation functions

```python
class Neuron:
    def __init__(self, num_inputs,activation):
        self.weights=torch.randn(num_inputs) * .1
        self.bias=torch.tensor(0.0)
        self.activation_name=activation

    def active(self,x):
        if self.activation_name=="sigmoid":
            return 1/1+torch.exp(-x)
        elif self.activation_name=="relu":
            return torch.maximum(torch.tensor(0.0), x)
        elif self.activation_name=="tanh":
            return torch.tanh(x)
        else:
            return x #linear activation (no activation)
        
    def forward(self, inputs):
        weighted_sum=torch.dot(self.weights,inputs)+self.bias
        output=self.active(weighted_sum)
        return output
```

```python
activations=['sigmoid', 'relu', 'tanh', 'linear']
for act in activations:
    neuron=Neuron(num_inputs=2, activation=act)
    print(f"Testing activation: {act}")
   # neuron.weights=torch.tensor([0.5,0.3]) #manually set weights for testing
    neuron.bias=.1

    inputs=torch.tensor([2.0,3.0]) #input values for the neuron
    output=neuron.forward(inputs)
    print(f"Weights: {neuron.weights}, Bias: {neuron.bias}")
    print(f"Output: {output:.4f}")
    print()
print("Great! We have successfully implemented a single neuron with multiple activation functions!")
```
